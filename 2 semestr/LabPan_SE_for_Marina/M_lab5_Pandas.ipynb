# Задание 1

```python
import pandas as pd
data = pd.read_csv("AppleStore.csv")

max = data.select_dtypes(include=['int64','float64']).max()
min = data.select_dtypes(include=['int64','float64']).min()
mean = data.select_dtypes(include=['int64','float64']).mean()
mid = data.select_dtypes(include=['int64','float64']).median()
print('\n\nМаксимальные значеия\n\n',max,'\n\nМинимальные значения\n\n',min,'\n\nСреднее арифмитическое значений\n\n',mean,'\n\nМедиана значений\n\n',mid)
```

# Задание 2

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("AppleStore.csv")

nu = 3 # значения должны быть в диапазоне [0,1]
area = data.loc[data['user_rating'] >= nu,'user_rating']
data_sum = area.sum()
data_mean = area.mean()
data_mean2 = data.loc[data['user_rating_ver'] == nu,'user_rating_ver'].mean() #Среднее по верефецированным пользователям 
data_srt = area.sort_values()
data_srt_back = area.sort_values(ascending=False)
print('\nДанные по одному полю\n',area,'\nСумма по группам\n',data_sum,'\nСортировка по возрастанию\n',data_srt,'\nСортировка по убыванию\n',data_srt_back)

plt.figure()
plt.bar(['Пользователей','Верефецированных пользователей'],[data_mean,data_mean2])
plt.title('Гистограмма по средней оценке')
plt.show()
```

# Задание 3

```python
import pandas as pd
data = pd.read_csv("AppleStore.csv")

col = ['user_rating','rating_count_tot','user_rating_ver','rating_count_ver']
groupy = data.groupby(['prime_genre']) # фильтрация по жанрам
min_col = groupy[col].min()
max_col = groupy[col].max()
mean_col = groupy[col].mean()
mid_col = data[col].median()
print('\nМинимальное по столбцу\n',min_col,'\nМаксимальное по столбцу\n',max_col,'\nСреднее по столбцу\n',mean_col,'\nМедиана по столбцу\n',mid_col)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("AppleStore.csv")

# График рейтинга скачиваний
gr_data = data.groupby(['prime_genre'])['rating_count_tot']
mean = gr_data.mean().sort_values()
# print(mean)
plt.figure()
plt.bar(mean.index,mean.values)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
data = pd.read_csv("AppleStore.csv")



filt_dat = data[(data['vpp_lic'] == 1)&(data['user_rating'] >= 0)&(data['rating_count_tot'] <= 20)&(data['rating_count_tot'] >= 10)]
table_sp = filt_dat.pivot_table('user_rating','rating_count_tot','rating_count_ver',fill_value=0)
fig = plt.figure(figsize=(15,8))
ax = Axes3D(fig)
print(table_sp)
x_pos = np.arange(table_sp.shape[0])
y_pos = np.arange(table_sp.shape[1])
print(y_pos)
xpos, ypos = np.meshgrid(x_pos, y_pos)
xpos = xpos.flatten()
ypos = ypos.flatten()

zpos=np.zeros(table_sp.shape).flatten()

zpos=np.zeros(table_sp.shape).flatten()
# print(xpos,ypos,zpos)
dx = .8
dy = 0.8
dz = table_sp.values.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz)
ax.w_xaxis.set_ticks(x_pos+.4)
# ax.w_xaxis.set_ticklabels(table_sp.index)
#ax.w_yaxis.set_ticks(ypos+.5)
ax.w_yaxis.reset_ticks()
ax.w_yaxis.set_ticks(y_pos+.4)
# ax.w_yaxis.set_ticklabels(table_sp.columns)

ax.set_xlabel('Рейтинг')
ax.set_ylabel('Общай рейтинг скачиваний')
ax.set_zlabel('Общей рейтинг верефецированных скачиваний')

plt.show()
```
